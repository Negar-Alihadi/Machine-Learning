## Data 

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

<div style="direction:rtl">
سه مرحله ی زیر را در این بلوک اجرا کنید :
<div style="direction:rtl">
1.یک شی از کلاس figure بسازید و نام آن را fig بگذارید.
<div style="direction:rtl">
2.یک axes (محور مختصات) با اندازه های [0,0,1,1] به آن اضافه کنید.
<div style="direction:rtl">
3.در محوری که برای این شی ساختید، نمودار y بر حسب x را رسم نمایید و نام این نمودار را نام خودتان قرار دهید و همچنین برای هر کدام از محور ها هم نامشان را قرار دهید.

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
plt.title('Negar Alihadi')
plt.xlabel('X')
plt.ylabel('Y')
```

<div style="direction:rtl">
یک شی از کلاس figure بسازید و نام آن را fig بگذارید.
<div style="direction:rtl">
دو axes (محور مختصات) با اندازه های [0,0,1,1] و [0.2,0.5,0.2,0.2] به آن اضافه کنید.

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,0.2,0.2])
```

<div style="direction:rtl">
حالا نمودار y بر حسب x را در هر دو محور این شی رسم نمایید و برای هر دو لیبل محور ها را قرار دهید.
<div style="direction:rtl">
پ.ن: برای این که نتیجه را روی شی خود مشاهده کنید، در انتهای بلوک fig را قرار دهید یعنی شی خود را صدا کنید تا خود را نمایش دهد..

```python
#Code Here
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes1.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
axes2=fig.add_axes([0.2,0.5,0.2,0.2])
axes2.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')

```

<div style="direction:rtl">
یک شی از کلاس figure بسازید و نام آن را fig بگذارید.
<div style="direction:rtl">
دو axes (محور مختصات) با اندازه های [0,0,1,1] و [0.2,0.5,0.4,0.4] به آن اضافه کنید.

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes( [0.2,0.5,0.4,0.4])
```

<div style="direction:rtl">
در محور بزرگتر این شی، نمودار z بر حسب x و در محور کوچکتر، نمودار y را بر حسب x  رسم کنید.
<div style="direction:rtl">
نکته مهم : دقت کنید که برای نمودار y بر حسب x، تنها می خواهیم بخشی از این نمودار را رسم کنیم یا به نوعی روی آن زوم کنیم.
<div style="direction:rtl">
پ.ن: برای این که نتیجه را روی شی خود مشاهده کنید، در انتهای بلوک fig را قرار دهید یعنی شی خود را صدا کنید تا خود را نمایش دهد.

```python
#Code Here
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes1.plot(x,z)
plt.xlabel('X')
plt.ylabel('Z')
axes2=fig.add_axes( [0.2,0.5,0.4,0.4])
axes2.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
axes2.set_xlim([20,22])
axes2.set_ylim([30,50])
```

<div style="direction:rtl">
با استفاده از مفاهیم شی گرایی، یک ساب پلات 2 در 1 بسازید و آن را به شی و محورهای خود تعلق دهید. 

```python
fig,axes=plt.subplots(1,2)
```

<div style="direction:rtl">
حالا نمودار y بر حسب x و z بر حسب x را در این دو ساب پلات با شکل و رنگ زیر رسم کنید.
<div style="direction:rtl">
پ.ن: برای این که نتیجه را روی شی خود مشاهده کنید، در انتهای بلوک fig را قرار دهید یعنی شی خود را صدا کنید تا خود را نمایش دهد.

```python
#Code Here
axes[0].plot(x,y,'b--',lw=2)
axes[1].plot(x,z,'r',lw=2)
fig
```

<div style="direction:rtl">
حالا سعی کنید با استفاده از ویژگی ها (Attribute) ی اشیا در کلاس ساب پلات و تغییر آن از مقدار دیفالت، نمودار های بالا را در اندازه های زیر رسم کنید.
<div style="direction:rtl">
در واقع اگه بخوام به فارسی سهل بگم برید سراغ تغییر figsize.

```python
plt.subplot(1,2,1)
plt.plot(x,y,'b',lw=3)
plt.subplot(1,2,2)
plt.plot(x,z,'r--',lw=3)
plt.xlabel('X')
plt.ylabel('Z')
```

<div style="direction:rtl">
همونطور که دیدید سعی کردیم تو این تمرین بیشتر رو کدزنی متپلاتلیب با متد شی گرایی (ساخت یک شی fig از کلاس figure) تمرکز کنیم.
<div style="direction:rtl">
تمام مثال های بالا رو فقط با دو تا متد plt.plot() و plt.subplot() هم انجام بدیم.
